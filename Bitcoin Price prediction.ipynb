### Calculating RSI

```python
def calculate_rsi(prices, window=50):
    deltas = np.diff(prices)
    positive_deltas = deltas * (deltas > 0)
    negative_deltas = -deltas * (deltas < 0)

    avg_gain = np.mean(positive_deltas[:window])
    avg_loss = np.mean(negative_deltas[:window])
    
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    
    return rsi
```

```python
calculate_rsi(df['Close**'])
```

### Calculating Bollinger Bands

```python
def calculate_bollinger_bands(price, window=10, std_dev=2):
    """
    Calculate Bollinger Bands in a DataFrame.

    Args:
        df (DataFrame): Input DataFrame with a column containing the prices.
        window (int): Window size for the Bollinger Bands (default: 10).
        std_dev (int): Standard deviation multiplier (default: 2).

    Returns:
        DataFrame: DataFrame with Bollinger Bands columns.

    """

    # Calculate the rolling mean and standard deviation
    rolling_mean = price.rolling(window).mean()
    rolling_std = price.rolling(window).std()

    # Calculate the upper and lower Bollinger Bands
    upper_band = rolling_mean + (rolling_std * std_dev)
    lower_band = rolling_mean - (rolling_std * std_dev)

    df['Upper Band'] = upper_band
    df['Lower Band'] = lower_band
    # Create a new DataFrame to store the Bollinger Bands
    bollinger_bands = pd.DataFrame({
        'Close': price,
        'MA': rolling_mean,
        'Upper Band': upper_band,
        'Lower Band': lower_band
    })

    return bollinger_bands
```

### Calculating Stochastic Oscillator

```python
def calculate_stochastic_oscillator(df, window=10):
    """
    Calculate Stochastic Oscillator in a DataFrame.

    Args:
        df (DataFrame): Input DataFrame with high, low, and close prices.
        window (int): Window size for the Stochastic Oscillator (default: 10).

    Returns:
        DataFrame: DataFrame with Stochastic Oscillator values.

    """

    # Calculate the lowest low and highest high over the window
    lowest_low = df['Low'].rolling(window).min()
    highest_high = df['High'].rolling(window).max()

    # Calculate the %K and %D values of the Stochastic Oscillator
    k_values = 100 * ((df['Close**'] - lowest_low) / (highest_high - lowest_low))
    d_values = k_values.rolling(3).mean()

    df['%K'] = k_values
    df['%D'] = d_values

```

### Calculating volume weighted average price (VWAP)

```python
df1 = pd.DataFrame()
```

```python
def calculate_vwap(df):
    """
    Calculate Volume Weighted Average Price (VWAP) in a DataFrame.

    Args:
        df (DataFrame): Input DataFrame with 'Volume', 'Close', and 'Typical Price' columns.

    Returns:
        float: Volume Weighted Average Price (VWAP) value.

    """

    # Calculate the typical price
    df1['Typical Price'] = (df['High'] + df['Low'] + df['Close**']) / 3

    # Calculate the product of typical price and volume
    df1['Typical Price x Volume'] = df1['Typical Price'] * df['Volume']

    # Calculate the cumulative sum of typical price x volume
    cumulative_sum = df1['Typical Price x Volume'].cumsum()

    # Calculate the cumulative sum of volume
    cumulative_volume = df['Volume'].cumsum()

    # Calculate the VWAP
    vwap = cumulative_sum / cumulative_volume

    # Return the last VWAP value
    return vwap.iloc[-1]
```

```python
calculate_vwap(df)
```

### Calculating Average True Range (ATR)

```python
def calculate_atr(df, window=10):
    """
    Calculate Average True Range (ATR) in a DataFrame.

    Args:
        df (DataFrame): Input DataFrame with 'High', 'Low', and 'Close' prices.
        window (int): Window size for the ATR calculation (default: 10).

    Returns:
        DataFrame: DataFrame with ATR values.

    """

    # Calculate the True Range (TR)
    df1['H-L'] = df['High'] - df['Low']
    df1['H-Cp'] = abs(df['High'] - df['Close**'].shift(1))
    df1['L-Cp'] = abs(df['Low'] - df['Close**'].shift(1))
    df1['TR'] = df1[['H-L', 'H-Cp', 'L-Cp']].max(axis=1)

    # Calculate the Average True Range (ATR)
    atr = df1['TR'].rolling(window).mean()

    # Create a new DataFrame to store the ATR values
    df['ATR'] = pd.DataFrame({'ATR': atr})
```

```python
calculate_atr(df)
```
